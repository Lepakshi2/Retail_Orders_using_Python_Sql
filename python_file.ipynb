```python
#importing the python libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#read data from the file and handle null values
ord_df = pd.read_csv(r"C:\Users\hp\Downloads\archive (2).zip",na_values=["unknown","Not Available"])
ord_df
```

```python
#fetch the values in ship mode coloumn which has unique
ord_df['Ship Mode'].unique()
```

```python
#rename columns names ..make them lower case and replace space with underscore
ord_df.rename(columns={'Order Id':'order_id', 'City':'city'})
ord_df.columns=ord_df.columns.str.lower()
ord_df.columns=ord_df.columns.str.replace(' ','_')
ord_df.head(5)
```

```python
# derive new columns discount,sales price and profit

ord_df['discount'] = ord_df['list_price'] * ord_df['discount_percent']*.01
ord_df['sales price'] = ord_df['list_price']-ord_df['discount']
ord_df['profit']= ord_df['sales price']-ord_df['cost_price']
```
